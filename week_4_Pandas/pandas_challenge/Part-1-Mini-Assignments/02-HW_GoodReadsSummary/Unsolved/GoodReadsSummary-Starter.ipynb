```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Path and filename for the file to Load
filePath = os.path.join('Resources','books_clean.csv')

# Read the modified GoodReads csv and store into Pandas DataFrame
gr_df = pd.read_csv(filePath)
gr_df.head()
```

```python
# Calculate the number of unique authors in the DataFrame

gr_authors_df = gr_df['Authors'].unique()
gr_authors_count = len(gr_authors_df)
# Calculate the earliest/latest year a book was published
gr_earliest_pub = gr_df['Publication Year'].min()
gr_latest_pub = gr_df['Publication Year'].max()

print(f"The earliest publication: {gr_earliest_pub}")
print(f"The latest publication: {gr_latest_pub}")
# Calculate the total reviews column that contains the sum of reviews (total - all authors)

gr_totals = gr_df[['Authors','One Star Reviews','Two Star Reviews','Three Star Reviews','Four Star Reviews','Five Star Reviews']].sum(axis='columns')
# Hint: use the pandas' sum(axis='columns') to get the sum across the columns 
#4942365
```

```python
# Create a `total_reviews` variable that sums up the values from your new `Total Reviews` column in the DataFrame
total_reviews = gr_totals.sum()

#total_reviews = gr_df['Totals'].sum()
print(f"Total reviews : {total_reviews}")
```

```python
# Place all of the data found into a summary DataFrame

gr_df['Totals'] = gr_totals


# Show the DataFrame
print(f"Total reviews : {total_reviews}")
gr_df

```
