```python
# Import Dependencies
import pandas as pd
```

```python
# Create reference to CSV file
csv_path = "Resources/Soccer2018Data.csv"

# Import the CSV into a pandas DataFrame
soccer_2018_df = pd.read_csv(csv_path, low_memory=False)
soccer_2018_df.head(3)
```

```python
# Collect a list of all the unique values in "Preferred Position"
footy_hlgn_pos = soccer_2018_df['Preferred Position'].unique()
footy_hlgn_pos
```

```python
# Looking only at strikers (ST) to start
bad_footy_df = soccer_2018_df.loc[soccer_2018_df['Preferred Position'] == "ST"]
bad_footy_df.head()
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
bad_footy_df = bad_footy_df.sort_values('ST')
bad_footy_df.head()
```

```python
# Reset the index so that the index is now based on the sorting locations
bad_footy_df = bad_footy_df.reset_index(drop=True)
bad_footy_df.head()
```

```python
# Save all of the information collected on the worst striker
worst_foot = bad_footy_df.loc[bad_footy_df['ST'].min() = bad_footy_df['ST'].min()]
worst_foot
```
